```python
import numpy as np 
import matplotlib.pyplot as plt
import h5py
```

```python
with h5py.File('corr_coef.h5', 'r') as file:
    variant = np.array(file['variant'])
    rm1z = np.abs(np.array(file['rm1z']))
    rm2z = np.abs(np.array(file['rm2z']))
file.close()
```

```python
variant
```

```python
X = ['0.25', '0.5', '0.75', '2', '5'] 
Ym1z = [rm1z[1], rm1z[2], rm1z[3], rm1z[4], rm1z[5]] 
Zm2z = [rm2z[1], rm2z[2], rm2z[3], rm2z[4], rm2z[5]] 

X_axis = np.arange(len(X)) 

plt.bar(X_axis - 0.2, Ym1z, 0.4, label = r'Correlation between $m_1$ and $z$', color='#AEC7E8') 
plt.bar(X_axis + 0.2, Zm2z, 0.4, label = r'Correlation between $m_2$ and $z$', color='#98DF8A')
plt.axhline(rm1z[0], color='#1F77B4')
plt.axhline(rm2z[0], color='#2CA02C')

plt.text(2.25, 0.0018, r'$m_1-z$ Correlation for Fiducial', fontsize=12, ha='right')
plt.text(2.25, 0.0027, r'$m_2-z$ Correlation for Fiducial', fontsize=12, ha='right')

plt.xticks(X_axis, X)
plt.xlabel(r"Common Envelope Efficieny ($\alpha$)") 
plt.ylabel("Magnitude of Mass-Redshift Correlation") 
plt.title("Variation with Common Envelope Efficiency") 
plt.legend() 
plt.show()
plt.savefig('CEa.pdf')
```
