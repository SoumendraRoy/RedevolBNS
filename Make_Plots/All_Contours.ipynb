```python
import numpy as np 
import matplotlib.pyplot as plt
import h5py
import corner
```

# All Contours

```python
def contours(var, informed_name):
    with h5py.File("/mnt/home/sroy1/Data_Release/compare_pop.h5", "r") as file:
        m1 = np.array(file["m1_"+var])
        m2 = np.array(file["m2_"+var])
        z = np.array(file["z_"+var])
    file.close()

    plt.figure()
    fig = corner.corner(np.array([m1, m2, z]).T, labels=[r'Primary Mass $(m_1)$',
                    r'Secondary Mass $(m_2)$', r'Redshift $(z)$'], color='brown')
    _ = fig.suptitle(informed_name)
    plt.savefig("Extra_Plots/"+var+".pdf")
```

```python
variant = ['N1e7_Fiducial_AllDCO_AIS', 'N1e7_CEa025_AllDCO_AIS', 'N1e7_CEa05_AllDCO_AIS', 'N1e7_CEa075_AllDCO_AIS',
           'N1e7_FRYER2022_AllDCO_AIS', 'N1e7_MullerMandel_AllDCO_AIS', 'N1e7_zetaHG5_AllDCO_AIS', 'N1e7_zetaHG5d5_AllDCO_AIS',
           'N1e7_zetaHG6_AllDCO_AIS', 'N1e7_beta025_AllDCO_AIS', 'N1e7_beta05_AllDCO_AIS', 'N1e7_beta075_AllDCO_AIS',
           'N1e7_beta1_AllDCO_AIS', 'N1e7_ECSN10_AllDCO_AIS', 'N1e7_ECSN200_AllDCO_AIS']

informed_names = ['Fiducial', r'CE$\alpha=0.25$', r'CE$\alpha=0.5$', r'CE$\alpha=0.75$', 'FRYER2022',
                  'MandelMüller ', r'$\zeta_{HG}=5$', r'$\zeta_{HG}=5.5$', r'$\zeta_{HG}=6$', r'$\beta=0.25$',
                  r'$\beta=0.5$', r'$\beta=0.75$', r'$\beta=1$', r'$\sigma_{ECSN}=10$', r'$\sigma_{ECSN}=200$']

for i in range(len(variant)):
    var = variant[i]
    informed_name = informed_names[i]
    print(var)
    contours(var, informed_name)
```
